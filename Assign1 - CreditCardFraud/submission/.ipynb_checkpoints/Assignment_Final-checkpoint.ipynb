 - WhiteBox : AdaBoost Classifier
 - BlackBox : Ensemble Classifier (with 'soft' voting)

```python
import numpy as np
import pandas as pd
from src.classifiers import *
% matplotlib inline

if __name__ == "__main__":
    # ------------ Step1 --------------- #
    filename = 'data/data_for_student_case_preprocessed_codes_basis.csv'
    df       = pd.read_csv(filename)
    data     = df.as_matrix()
    X_orig   = data[:,:-1]
    Y_orig   = data[:,-1].astype(int)
    rand_idx = np.random.choice(len(X_orig), len(X_orig), replace=False)
    X_orig   = X_orig[rand_idx]
    Y_orig   = Y_orig[rand_idx]
    print (' - Original Y : ', Counter(Y_orig), ' || Type : ', Y_orig.dtype)
    
    # -------------- Step2 ------------- #
    args = {  'random_state'          : 25,
              'sampling_ratio'        : 0.3,
              'LogisticRegression'    : {'bool':1, 'C':1, 'max_iter':500}
            , 'RandomForestClassifier': {'bool':0, 'n_estimators' : 200}
            , 'AdaBoostClassifier'    : {'bool':0, 'n_estimators': 200, 'learning_rate':1.0}
            , 'SVC'                   : {'bool':0}
            , 'XGBClassifier'         : {'bool':0}
            , 'ensemble'              : {'bool':0}
           }
    smote  = 1
    cv     = 5
    print (' -------- PARAMS ----------- ')
    print (' -- filename : ', filename)
    print (' -- Total Features : ', len(df.columns) - 1)
    print (' -- Sampling Ratio : ', args['sampling_ratio'])
    
    classifiers, conf_matrixes, Y_tests, Y_tests_preds, Y_tests_preds_probabs = \
                experiments(X_orig, Y_orig, cv, smote, args)
    
    if (args['RandomForestClassifier']['bool'] or args['AdaBoostClassifier']['bool'] or args['XGBClassifier']['bool']):
        clf    = classifiers[0] # if RandomForests
        coeffs = clf.feature_importances_

        if args['RandomForestClassifier']['bool'] == 1:
          plt.title('RandomForest - Feature Importances')
        elif args['AdaBoostClassifier']['bool']:
          plt.title('AdaBoost - Feature Importances')
        plt.bar(range(len(coeffs)), coeffs)
        plt.xticks(range(len(df.columns[:-1])), df.columns[:-1])
        plt.xticks(rotation=90)
```
